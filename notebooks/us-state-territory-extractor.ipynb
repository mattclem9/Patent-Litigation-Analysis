```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re as re
```

```python
from pathlib import Path
ROOT = Path().resolve().parent  # project root
```

```python
cases=pd.read_csv(ROOT / 'data/processed/cases/cases_cleaned_v3.csv')
```

```python
print(f" {len(cases)}")
```

```python
US_STATES = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado",
    "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Idaho",
    "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana",
    "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota",
    "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada",
    "New Hampshire", "New Jersey", "New Mexico", "New York",
    "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon",
    "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota",
    "Tennessee", "Texas", "Utah", "Vermont", "West Virginia", "Washington",
    "Virginia", "Wisconsin", "Wyoming", "District of Columbia", "Puerto Rico", "Virgin Islands", "Guam"
]

def extract_state(court_name):
    if pd.isna(court_name):
        return None
    if re.search(r'dc district', str(court_name), re.IGNORECASE):
        return 'District of Columbia'
    if re.search(r'virgini islands', str(court_name), re.IGNORECASE):
        return 'Virgin Islands'
    for state in US_STATES:
        if re.search(state, str(court_name), re.IGNORECASE):
            return state
    return None

cases["U.S. State/Territory"] = cases["court_name"].apply(extract_state)

print(cases["U.S. State/Territory"].value_counts())
print("\nUnmatched rows:")
print(cases[cases["U.S. State/Territory"].isna()]["court_name"].unique())
```

```python
print(f"Total rows: {len(cases)}")
print(f"State/Territory filled: {cases['U.S. State/Territory'].notna().sum()}")
```
