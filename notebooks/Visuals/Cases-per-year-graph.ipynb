```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re as re
import seaborn as sns
import plotly.express as px
```

```python
from pathlib import Path
ROOT = Path().resolve().parent.parent  # project root
```

```python
cases=pd.read_csv(ROOT / 'data/processed/cases/cases_cleaned_v2.csv')
```

```python
cases['date_filed'] = pd.to_datetime(cases['date_filed'], errors='coerce')
cases['year'] = cases['date_filed'].dt.year
cases_per_year = (
    cases
    .groupby('year')
    .size()
    .reset_index(name='num_cases')
)
```

```python
cases_per_year = cases_per_year[
    ~cases_per_year['year'].isin([1963, 2021])
]
```

```python


cases['date_filed'] = pd.to_datetime(cases['date_filed'], errors='coerce')
cases['year'] = cases['date_filed'].dt.year

cases_per_year = (
    cases
    .groupby('year')
    .size()
    .reset_index(name='num_cases')
)


cases_per_year = cases_per_year[~cases_per_year['year'].isin([1963, 2021])]


fig = px.line(
    cases_per_year,
    x='year',
    y='num_cases',
    title='Patent Litigation Cases Filed from 1972-2020'
)
fig.update_traces(line=dict(color='blue', width=2.5))


fig.add_vline(
    x=1982,
    line_width=2.5,
    line_dash="solid",
    line_color="red",  
    annotation_text="Federal Circuit (1982)",
    annotation_position="top left",
    annotation_x=1981.5
)

fig.add_vline(
    x=2011,
    line_width=2.5,
    line_dash="solid",
    line_color="green",  
    annotation_text="America Invents Act (2011)",
    annotation_position="top left",
    annotation_x=2010.5
)



fig.update_layout(
    title=dict(
        text="<i>Patent Litigation Cases Filed from 1972-2020</i>",
        x=0.5,             
        xanchor="center",
        font=dict(
            size=18,
            family="Arial",
            color="black"
        )
    ),
    xaxis_title="Year of Date Filed",
    yaxis_title="Number of Cases Filed",
    template="simple_white"
)



fig.update_yaxes(tickformat=",")  




fig.show()
```
