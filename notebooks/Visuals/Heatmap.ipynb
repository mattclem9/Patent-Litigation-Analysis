```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re as re
import plotly.express as px
```

```python
from pathlib import Path
ROOT = Path().resolve().parent.parent  # project root
```

```python
cases=pd.read_csv(ROOT / 'data/processed/cases/cases_cleaned_v4.csv')
```

```python
state_abbrev = {
    "Alabama": "AL", "Alaska": "AK", "Arizona": "AZ", "Arkansas": "AR",
    "California": "CA", "Colorado": "CO", "Connecticut": "CT", "Delaware": "DE",
    "Florida": "FL", "Georgia": "GA", "Hawaii": "HI", "Idaho": "ID",
    "Illinois": "IL", "Indiana": "IN", "Iowa": "IA", "Kansas": "KS",
    "Kentucky": "KY", "Louisiana": "LA", "Maine": "ME", "Maryland": "MD",
    "Massachusetts": "MA", "Michigan": "MI", "Minnesota": "MN", "Mississippi": "MS",
    "Missouri": "MO", "Montana": "MT", "Nebraska": "NE", "Nevada": "NV",
    "New Hampshire": "NH", "New Jersey": "NJ", "New Mexico": "NM", "New York": "NY",
    "North Carolina": "NC", "North Dakota": "ND", "Ohio": "OH", "Oklahoma": "OK",
    "Oregon": "OR", "Pennsylvania": "PA", "Rhode Island": "RI", "South Carolina": "SC",
    "South Dakota": "SD", "Tennessee": "TN", "Texas": "TX", "Utah": "UT",
    "Vermont": "VT", "Virginia": "VA", "Washington": "WA", "West Virginia": "WV",
    "Wisconsin": "WI", "Wyoming": "WY", "District of Columbia": "DC",
    "Puerto Rico": "PR", "Guam": "GU", "Virgin Islands": "VI",
    "American Samoa": "AS", "Northern Mariana Islands": "MP"
}
state_counts = cases.groupby("U.S. State/Territory").size().reset_index(name="cases")
state_counts["abbrev"] = state_counts["U.S. State/Territory"].map(state_abbrev)
```

```python
state_counts["log_cases"] = np.log1p(state_counts["cases"])

fig = px.choropleth(
    state_counts,
    locations="U.S. State/Territory",  
    locationmode="USA-states",
    color="cases",
    scope="usa",
    color_continuous_scale="Reds",
    title="Patent Litigation Cases by State 1972-2020",
    hover_data={"cases": True, "log_cases": False}
)

fig.write_html("cases_by_state.html")
```
